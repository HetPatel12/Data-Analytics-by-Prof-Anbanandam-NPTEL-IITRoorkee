```python
# Test the hypothesis that "gender and student motivation" are independent
obs = pd.pivot_table(acad[['g','sm']], index='g', columns='sm', aggfunc=len)
obs
#aggfunc is aggregate function
```

```python
# perform chi2 test to check independence
from scipy.stats import chi2_contingency
```

```python
chi2, p, dof, tbl = chi2_contingency(obs)
```

```python
chi2 # chi2 value
```

```python
p # p-value, more than 0.05 so accept null hypothesis 
# by accepting null hypothesis we conclude gender and student motivation are independent
```

# Goodness Of Fit Test

degrees of freedom = k-1-p, k: no. of categories/observations, p:no. of parameters estimated from the sample data

Uniform Distribution: parameter=0

Poisson Distribution: parameter=1 (lambda value)

Normal Distribution: parameter=2 (mean and variance)

```python
import scipy
from scipy.stats import chi2
from scipy.stats import poisson

import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('P_distribution.xlsx')
data
```

```python
Observed_Freq = data['Frequency']
```

```python
total_arrival = 600
total_time_period = 100
mu = total_arrival/total_time_period
```

```python
Expected_Freq = []
for i in range(len(Observed_Freq)):
    E_Freq = 100*poisson.pmf(i,mu)
    Expected_Freq.append(E_Freq)

Expected_Freq
# for each x-value we got our expected frequency value using for loop
```

```python
Expected_Freq_round_off = [round(elem, 2) for elem in Expected_Freq]
Expected_Freq_round_off
```

```python
df = pd.DataFrame(list(zip(Observed_Freq, Expected_Freq_round_off)), columns = ['Observed Frequency','Expected Frequency'])
df
```
