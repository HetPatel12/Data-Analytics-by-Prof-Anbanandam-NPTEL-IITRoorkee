```python
# it might give wrong output if my excel sheet is wrong
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
import statsmodels as sm
```

```python
df = pd.read_excel('twoway.xlsx')
df
```

```python
data = pd.melt(df.reset_index(), id_vars=['index'], value_vars=['Prep_pro','Business','Engineering','Artsandscience'])
data.columns = ['index','treatments','value']
data
```

```python
model = ols('Value ~ C(Business)+C(Engineering)+C(Artsandscience)+C(prep_pro)+C(Business)+C(Engineering)+C(Artsandscience): C(prep_pro)', data=df).fit()
aov_table = sm.stats.anova_lm(model, typ=2)
print(aov_table)
```

```python
formula = 'Value ~ C(Business)+C(Engineering)+C(Artsandscience)+C(prep_pro)+C(Business)+C(Engineering)+C(Arts & Sciences): C(prep_pro)'
model = ols(formula, df).fit()
aov_table = sm.stats.anova_lm(model,typ=2)
# typ=2 means 2 anova tables
print(aov_table)
```
